```python
def file_read(fname):
        content_array = []
        with open(fname,encoding='utf-8') as f:
                #Content_list is the list that contains the read lines.     
                for line in f:
                        line = line.replace('\n', '')
                        content_array.append(line)
                return content_array

array = file_read('urls.txt')
itemSkip = 0
array = array[itemSkip:]
```

```python
import os
import re
import unicodedata
from unicodedata import*
name = 'vov-thegioi'
def makeTxt(text,index,name):
  filename = "./dataset/"+name+"-"+str(index)+".txt"
  dirname = os.path.dirname(filename)
  if not os.path.exists(dirname):
      os.makedirs(dirname)
  with open(filename, 'w', encoding='utf-8') as f:
    f = open(filename, "a", encoding='utf-8')
    f.write(text)
    f.close()
def url_get(link,index):
  url_page= link
  page = requests.get(url_page)
  soup = BeautifulSoup(page.content,'html.parser')
  main = soup.find('div',class_= 'maincontent')
  
  if(main == None):
    return False
  divs = main.findAll('p')
  if(divs == None):
    return False
  texts = ""
  for div in soup.findAll("p", {'class':"t-c"}): 
    if(div == None):
      continue
    div.decompose()
  for div in divs:
    # link = div.find('a').get('href')
    text = div.find(text=True)
    if(text == None):
      continue
    if(texts == ""):
      texts = text
      continue
    texts = texts +" "+ text
  texts= texts_cleaning(texts)
  makeTxt(texts,index,name)
  return True
def url_get_thanhnien(link,index):
  url_page= link
  page = requests.get(url_page)
  soup = BeautifulSoup(page.content,'html.parser')
  main = soup.find('div',class_= 'cms-body detail')
  if(main == None):
    return False
  divs = main.findAll('p')
  if(divs == None):
    return False
  texts = ""
  for div in divs:
    text = div.text
    if(text == None):
      continue
    if(texts == ""):
      texts = texts  + text
      continue
    texts = texts +" "+ text
    texts = re.sub("\s\s+" , " ", texts)
  texts=texts_cleaning(texts)
  makeTxt(texts,index,name)
  return True

def url_get_vov(link,index):
  url_page= link
  try:
    page = requests.get(url_page)
  except:
    return False
  soup = BeautifulSoup(page.content,'html.parser')
  main = soup.find('div',class_= 'article-content')
  if(main == None):
    return False
  divs = main.findAll('p')
  if(divs == None):
    return False
  texts = ""
  for div in divs:
    text = div.text
    if(text == None):
      continue
    if(texts == ""):
      texts = texts  + text
      continue
    texts = texts +" "+ text
    texts = re.sub("\s\s+" , " ", texts)
    texts = texts.replace('./.','.')
  texts =texts_cleaning(texts)
  makeTxt(texts,index,name)
  return True

def texts_cleaning(texts):
  texts = unicodedata.normalize('NFKC', texts)
  texts = unicodedata.normalize('NFKD', texts)
  for i in range (100):
    numerator, _slash, denominator = texts.partition('⁄')
    if(len(denominator)!=0):
      texts = numerator+"/"+denominator
    if(len(denominator)==0):
      break
  texts = re.sub(r'^https?:\/\/.*[\r\n]*', '', texts, flags=re.MULTILINE)
  texts = re.sub("[\(\[].*?[\)\]]", "", texts)
  texts = texts.strip()
  texts = re.sub("\s\s+" , " ", texts)
  chars = "!@#$%^&*_+-={}[]|\?><~`"
  for i in range(len(chars)):         #can't use len() alone b/c int obj is not iterable
      texts = texts.replace(chars[i], "")
 
  return texts
```

```python
def fraction_finder(s):
    for c in s:
        try:
            name = unicodedata.name(c)
        except ValueError:
            continue
        if name.startswith('VULGAR FRACTION'):
            normalized = unicodedata.normalize('NFKC', c)
            numerator, _slash, denominator = normalized.partition('⁄')
            yield c, int(numerator), int(denominator)
```
